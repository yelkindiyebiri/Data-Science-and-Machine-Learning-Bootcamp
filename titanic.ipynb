```python
import pandas as pd
data = pd.read_csv("titanic_train.csv")
data.head()
```

```python
import numpy as np
import pandas as pd
survived_sex = pd.crosstab(index=data['Survived'], columns = data['Sex'])

survived_sex.index = ['died','survived']
survived_sex
```

```python
import numpy as np
import pandas as pd
survived_sex = pd.crosstab(index=data['Survived'], columns = data['Pclass'])
survived_sex.index = ['died', 'Pclass']
survived_sex.columns = ['class1', 'class2', 'class3']
survived_sex
```

```python
char_cabin = data['Cabin'].astype(str) #convert cabin to str

new_Cabin = np.array([cabin[0] for cabin in char_cabin]) #take first letter
data["Cabin"] = pd.Categorical(new_Cabin)  # save cabin 
data.head()
```

```python
my_tab = pd.crosstab(index = data["Survived"], columns ="count")
my_tab
```

```python
pd.crosstab(index=data["Cabin"], columns="count")
```

```python
pd.crosstab(index=data["Cabin"], columns=data["Survived"])
```

```python
#Table of survival vs passenger class
survived_class = pd.crosstab(index=data["Survived"], columns=data["Pclass"], margins= True)
survived_class
```

```python
survived_class.columns = ["class1", "class2", "class3", "rowtotal"]
survived_class.index = ["died", "survived", "coltotal"]

survived_class
survived_class/survived_class.loc["coltotal","rowtotal"]
```
