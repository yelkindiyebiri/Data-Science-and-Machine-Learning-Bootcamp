## Giriş

Sınıflandırma problemleri temelde belli bir girdi verisinden etiket belirlemeye dayanır. Bu bağlamda yöntemler

$$
Y = C(X) \text{, } Y \in C=\left \{C_1, C_2, ..., C_K \right \}
$$

şeklinde belirtilen $C(x)$ sınıflandırma fonksiyonunu kestirmeye yöneliktir. Burada $X$ etiketin ne olduğuna ilişkin belirleyici öznitelikleri (feature) barındıran bir vektörür. Bu vektör $Y$ sınıf etiketini üretmek için bağımlı değişkenleri ifade eder. Lineer regresyondan farklı olarak kestirilen sınıflandırma fonksiyonu $C$ kümesinden genelde sıralı olamayan etiketleri kestirmeyi sağlar. Daha önce Lineer Regresyonda üzerinde çalıştığımız problemlerde kestirilen fonksiyon sıralı ve genelde sürekli kuantitatif (quantitative) değerleri kestirmek için kullanılıyordu. Sınıflandırmada ise $C$ kümesinden kualitatif (nitel) sınıf etiketi ($Y$) belirlemek için yöntemleri inceleyeceğiz.

Kalitatif değişkenler genelde sıralı olmayan bir değer kümesinden değerler alırlar. Uydu görüntülerinden arazi tipi:

$$
C = \left \{tarla, orman, yerleşim yeri, sulak arazi \right \}
$$

Ya da tarlada ekili olan ürün tipi:

$$
C = \left \{buğday, mısır, nohut, fasulye\right \}
$$

örnekler verilebilir. Genellikle sınıflandırmada belli bir öznitelik vektörü verildiğinde bu örneklemin hangi sınıfa ait olabileceği ile ilgili olasılık değerleri ile ilgileniriz. Bu olasılık değerleri bazı durumlarda sınıf etiketinden daha da fazla önem taşır. Örneğin belli bir bölgede doğal doğal afet yaşanma olasılığı o bölgenin doğal afet bölgesi olarak sınıflandırılmasından daha fazla bilgi içerir. Sonuçta olasılıklar belirlendirildiğinde sınıflandırma genel olarak belli bir eşik değerin seçilmesi sonrası rahatlıkla yapılabilir.

Sınıflandırma yöntemleri genel olarak girdi uzayını ($X$) belli bölgelere (region) ayırarak o bölgelerin etiketini belirlemeye dayalıdır. Bu bölgelerin sınırlarına karar sınırları (decision boundary) diyoruz. Karar sınırları üzerinde iken $C(X)$ fonksiyonu bu sınıra komşu olan her sınıf için aynı olasılık değerini vermektedir. Dolayısı ile, karar sınırlarını oluşturan fonksiyonları kestirmek sınıflandırma yöntemlerinde genellikle temel amaçtır.

Her sınıf etiketi, $k$, için indikatör (belirteç) fonksiyonlarımız kestirilmiş olsun, $\hat{d_k}(X)$, olduğunu varsayalım. Ayrıca bu indikatör(belirteç) fonksiyonlarının aşağıda verildiği gibi olduğunu düşünelim:

$$
\hat{d_k}(X) = \hat{\beta_{k0}} + \hat{\beta_k}^TX 
$$

Görüldüğü gibi indikatör fonksiyonu girdi vektörü (öznitelik vektörü) açısından doğrusal bir fonksiyondur. Bu indikatör fonksiyonlarını kullanarak sınıflandırma fonksiyonu $C(X)$ şu şekilde tanımlanabilir.

$$
C(X) = \underset{k}{\mathrm{argmin}} \text{  } d_k(X)
$$

Bir başka değişle $C(X)$ fonksiyonunun ($1,2,3,..,k,..,K$) şeklinde ifade edilen sınıf etiketi değerlerinden hangisinin indikatör fonksiyonu en büyük ise onun indeksini ($k$) döner. Bu durumda iki farklı sınıf için karar sınırı (KS) aslında her ikisinin de indikatör fonksiyonunun aynı değeri ürettiği öznitelik değerleri kümesi (noktalar, X) olacaktır.

$$
KS = d_k(X) - d_l(X) =  \left \{ X | (\hat{\beta_{k0}}-\hat{\beta_{l0}}) + (\hat{\beta_{k}}-\hat{\beta_{l}})^TX = 0\right \}
$$

Bir başka değişle karar sınırı, belli bir uzaklık metriği açısından, her iki sınıfa da aynı uzaklıktadır. Yukarıdaki denkleme bakıldığında aslında karar sınırının bir hiper düzlem oluşturduğunu söyleyebiliriz. Bu düzlem normali $(\hat{\beta_{k}}-\hat{\beta_{l}})$ ve offset değeri $(\hat{\beta_{k0}}-\hat{\beta_{l0}})$ ile ifade edilmektedir.

**Q) Şekil üzerine çizilmis karar sınırının denklemi nedir?**


Bu problemi klasik regresyon yöntemi ile çözmek icin sınıflandırma problemini değişken dönüşümü yardımı ile şu şekilde ifade edelim. Kara listede olma durumu $Y$:

$$
Y = \begin{cases}
0 & \text {Eğer Hayırsa}\\
1 & \text {Evetse}
\end{cases}
$$

Bu durumda $Y=C(X)$ fonksiyonunu lineer regresyonla modelleyebiliriz.

$$
Y = \beta_0 + \beta_1 Income + \beta_1 Balance
$$

Eğer herhangi bir $X = (Income, Balance)$ degeri için kestirilen model 0.5 ten büyük bir değer verirse Kara liste kuçuk değer verirse Normal kullanıcı olarak sınıflandırabiliriz.

```python

import warnings
warnings.filterwarnings('ignore')
# Some part of the following code segment is Taken from https://github.com/JWarmenhoven/ISLR-python.git
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns

df = pd.read_excel('Default.xlsx')

# Note: factorize() returns two objects: a label array and an array with the unique values.
# We are only interested in the first object. 
df['default2'] = df.default.factorize()[0]
df['student2'] = df.student.factorize()[0]
df.head(5)

Y = df.default2.to_numpy()

X = np.vstack(
(
    np.ones(len(Y)),
    df.balance.to_numpy(),
    df.income.to_numpy()
)
).T


betas = np.linalg.lstsq(X,Y, rcond=None)[0]

print(betas)

#Tahmin modeli çıktıları

y = np.dot(X,betas)

print ('Maximum Y ', np.max(y))

y[y<0.5] = 0
y[y>=0.5] = 1

print (len(y[y==1]))

df['default3'] = y


fig = plt.figure(figsize=(7,12))
ax1 = plt.subplot(211)
ax1.scatter(df[df.default2==1].balance, df[df.default2==1].income, s=50, c='orange', marker='+',
            linewidths=1,alpha=0.5)
ax1.scatter(df[df.default2==0].balance, df[df.default2==0].income, s=50, marker='o', linewidths='1',
            edgecolors='lightblue', facecolors='white', alpha=.5)


##Burada tahminlerimizi de ayrı bir sembol olarak basalım
ax1.scatter(df[df.default3==0].balance, df[df.default3==0].income, s=5, c='red', marker='o',
            linewidths=1)

ax1.set_ylim(ymin=0)
ax1.set_ylabel('Income')
ax1.set_xlim(xmin=-100)
ax1.set_xlabel('Balance')

ax2 = plt.subplot(212)

sns.regplot(df.balance, df.default2, order=1, ci=None,
            scatter_kws={'color':'orange'},
            line_kws={'color':'lightblue', 'lw':2}, ax=ax2)

ax2.hlines(1, xmin=ax2.xaxis.get_data_interval()[0],
          xmax=ax2.xaxis.get_data_interval()[1], linestyles='dashed', lw=1)
ax2.hlines(0, xmin=ax2.xaxis.get_data_interval()[0],
          xmax=ax2.xaxis.get_data_interval()[1], linestyles='dashed', lw=1)
ax2.set_ylabel('Probability of default')
ax2.set_xlabel('Balance')
ax2.set_yticks([0, 0.25, 0.5, 0.75, 1.])
ax2.set_xlim(xmin=-100)


```

Şeklinde değişken dönüşümü yapmak istediğimizde sınıfların birbirine göre sıralaması ve sınıflar arası uzaklıkların birbirine eşit olması varsayımları bizi yanlış sınıflandırmaya kolayca götürebilir.



İndikatör fonksiyonunu koşullu olasılıktan faydalanarak belli bir öznitelik vektörü $(X = x)$ gözlemlendiğinde verinin $k$ sınıfında olma olasılığı şeklinde de yazabiliriz.

$$
d_k(x) = P(C = k | X = x)
$$

Bu ifade olasılık (burada posterior probability) ifade ettiğinden 0 ila 1 arasında değer alır. Eğer böyle bir olasılık fonksiyonu modellenebilir ise ve ayrıca bu fonksiyon ($d_k(x)$) veya bu fonksiyonun monoton transformasyonu (örneğin $\text{log }(d_k(x))$) girdi uzayına göre lineer ise karar sınırının doğrusal olduğu söylenebilir. Eğer yukarıdaki şekilde olduğu gibi bu olasılık fonksiyonunu doğrusal bir fonksiyon olarak yazar isek, fonksiyon negatif değerler alabilir. Ya da 1'den çok büyük değerler alabilir. Bu durumda şekilde de görüldüğü gibi sınıflandırmada sorunlara yol açabilir. Bu olasılık fonksiyonunu ifade edebilecek fonksiyonlar Logistik Regresyon, Lineer Diskriminant ve Quadratik Dscriminant analizi ile bulunabilir.

## Logistik Regresyon

İkili bir sınıflandırma probleminde (örneğin yukarıdaki kredi kartı kara liste problemi) indikatör fonksiyonumuzu posterior olasılığın bir ifadesi olarak yazabildiğimizi düşünelim.

Bu durumda gözlemlediğimiz değer aşağıdaki sınıflardan birinde olacaktır.

$$
C = \left \{ 0 \text{: Kara listede}, 1 \text{: Kara listede değil} \right \}
$$


Bu durumda $Y = C(x)$ sınıf etiketi aşağıdaki gibi gösterilebilir.
$$
Y = \begin{cases}
0 & \text {Eğer Kara Listede Değilse}\\
1 & \text {Kara Listede ise}
\end{cases}
$$

Belli bir kişi için gözlenen öznitelik vektörü $X = x = (Balance, Income)$ verildiğinde bu kişinin kara listede olma olasılığı koşullu olasılık olarak aşağıdaki gibi yazılabilir:

$$
d_k(x) = p(x) = P(Y = 1 | X = x)
$$

Logistik regresyonunda buradaki $p(x)$ olasılık fonksiyonu:

$$
p(x) = \frac{e^{\beta_0 + \beta^Tx}}{1+e^{\beta_0 + \beta^Tx}}
$$

şeklinde yazılır. Bu fonksiyon tanımı gereği $x$ hangi değeri alırsa alsın $[0,1]$ aralığında değer alır. Bu kapsamda koşullu olasılığı ifade edebileceği söğlenebilir. İki sınıflı yukarıdaki problemde indikatör fonksiyonlarını yazdığımızda:

kara liste olmama sınıfı için koşullu olasılık:
$$
d_1(x) = p_1(x) = P(Y = 1 | X = x) = \frac{e^{\beta_{10} + \beta_1^Tx}}{1+e^{\beta_{10} + \beta_1^Tx}}
$$

kara listede olmama durumu ($Y=0$) ise, sadece iki sınıfımız olduğu için ve herhangi bir kişi her iki sınıftan sadece birinde bulabileceği için:

$$
d_0(x) = p_0(x) = P(Y = 0 | X = x) = 1 - p_1(x) = \frac{1}{1+e^{\beta_{10} + \beta_1^Tx}}
$$

şeklinde yazılabilir. Bu durumda karar sınırı bu iki olasılığın eşit olduğu noktaları kapsar.

$$
d_1(x) = d_0(x)
$$

Bir başka ifade ile 

$$
\frac{d_1(x)}{d_0(x)} = 1
$$

Bu ifadede her iki tarafın logaritmasını aldığımızda 

$$
\text{log }(\frac{d_1(x)}{d_0(x)}) = \text{log}(1) = 0
$$

ifadesini elde ederiz. olasılık fonksiyonlarının tanımını yerine yazdığımızda

$$
\text{log }(\frac{d_1(x)}{d_0(x)}) = \text{log }(\frac{\frac{e^{\beta_{10} + \beta_1^Tx}}{1+e^{\beta_{10} + \beta_1^Tx}}}{\frac{1}{1+e^{\beta_{10} + \beta_1^Tx}}} = \text{log }(e^{\beta_{10} + \beta_1^Tx}) = \beta_{10} + \beta_1^Tx = 0
$$

ifadesi elde edilir. burada yapılan monoton transformasyona log odds veya logit transformasyonu denir. Denklemden göründüğü üzere, daha önce de bahsettiğimiz gibi karar sınırı monoton transformasyonunda yine doğrusal bir denklem olarak ifade edilmektedir. Ancak bu denklemin parametereleri $\beta$ lineer regresyon yöntemi ile değil, Logistik regresyon yöntemi ile bulunmaktadır.


```python
import sklearn.linear_model as skl_lm
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.metrics import confusion_matrix, classification_report, precision_score
from sklearn import preprocessing
from sklearn import neighbors

import statsmodels.api as sm
import statsmodels.formula.api as smf

X_train = df.balance.values.reshape(-1,1) 

y = df.default2

# Create array of test data. Calculate the classification probability
# and predicted classification.
X_test = np.arange(df.balance.min(), df.balance.max()).reshape(-1,1)

clf = skl_lm.LogisticRegression(solver='newton-cg')
clf.fit(X_train,y)
prob = clf.predict_proba(X_test)
ax = plt.subplot(111)
# Right plot
ax.scatter(X_train, y, color='orange')
ax.plot(X_test, prob[:,1], color='lightblue')

ax.hlines(1, xmin=ax.xaxis.get_data_interval()[0],
          xmax=ax.xaxis.get_data_interval()[1], linestyles='dashed', lw=1)
ax.hlines(0, xmin=ax.xaxis.get_data_interval()[0],
          xmax=ax.xaxis.get_data_interval()[1], linestyles='dashed', lw=1)
ax.set_ylabel('Probability of default')
ax.set_xlabel('Balance')
ax.set_yticks([0, 0.25, 0.5, 0.75, 1.])
ax.set_xlim(xmin=-100)
```

Bulunan karar sınır deklemi girdi uzayını ikiye bölen bir hiper düzlem oluşturur. Bu düzlemin normali $\beta_1$, offset değeri ise $\beta_{10}$ dir. Sınıflandırma için veriyi optimal olarak ikiye ayıran yöntemlerden biri de Rosenblatt tarafından önerilen Perceptron yöntemidir. Bu yöntemde ikili bir sınıflandırma için optimum hiper düzlem bir eşik fonksiyonu aracılığı ile belirlenir.

$$
d_1(x) = f(x) = \begin{cases}
1 & \text{eğer  } w^Tx + b > 0\\
0 & \text{değilse}
\end{cases}
$$

Burada $w$ ağırlıkları, $x$ sınıflandırma için kullanılacak olan öznitelikleri (feature), $b$ ise bias olarak ifade edilmektedir. Bu ifade eğer aktivasyon fonksiyonu step fonksiyonu olarak tanımlanırsa yapay sinir ağlarının en temel elementi olan single-layer perceptron olarak da görülebilir. Burada aktivason fonksiyonu sigmoid olduğunda logistik regresyonu ile elde edilen indikatör fonksiyonuna ne kadar benzediğini de görebilirsiniz.

Sınıfların net olarak ayrılabildiği problemlerde logistik regresyonu veya perceptron gibi hiper düzlem bulan yöntemler stabilizasyon sorunları yaşayabilir. ### Birden çok sınıf için logistik regresyonu

Yukarıda bahsedildiği gibi sınıflandırma yöntemlerinin birçoğu karar sınırlarını optimum olarak veya sınıflar arasında belli kesişimlere izin verek bulmaya çalısırlar. Birden çok sınıf olması durumunda her sınıf ikilisi arasındaki düzlemler bulunarak girdi uzayı birbirinden ayrılmış bölgelere ayrılabilir. Bu bölgeler karar sınırlarının doğrusal denklem olarak ifade edildiği sınırlar ile belirlenmiş olur.

$K$ farklı sınıftan oluşan bir sınıflandırma probleminde her bir sınıfın posterior olasılık fonksiyonu $p_k(x)$ şeklinde verilsin. Herhangi bir nokta için olasılıkların toplamının bir olması beklenir (neden ?). Bu durumun sağlanabilmesi için genellikle son sınıf referans seçilerek diğer sınıfların olasılıkları yazılır.

$$
p_k(x) = P(C=k | X=x) =  \frac{e^{\beta_{k0} + \beta_k^Tx}}{1+\sum_{l=1}^{K-1}e^{\beta_{l0} + \beta_l^Tx}}
$$

Referans sınıf için ise:
$$
p_K(x) = P(C=K | X=x) =  \frac{1}{1+\sum_{l=1}^{K-1}e^{\beta_{l0} + \beta_l^Tx}}
$$

şeklinde yazılabilir.


Bu durumda logit fonksiyonları:

$$
\text{log }(\frac{P(C=k|X=x)}{P(C=K|X=x)}) = \beta_{k0} + \beta_k^Tx
$$

şeklinde yazılabilir ve her bir sinifi ayiran duzlem denkleminin parametreleri görülebilir. Sınıflandırma probleminin parametreleri $\theta = \left \{ \beta_{10},\beta_1,  \beta_{20},\beta_2, ... , \beta_{k-1 0},\beta_{k-1} \right \}$ şeklinde bir vektör olarak ifade edilebilir. Bu parametre vektörünü kullanarak olasılık fonksiyonları 

$$
p_k(x|\theta) = P(C = k| X = x)
$$

şeklinde yazılır. Parametreler $\theta$ en yüksek olabilirlik (Maksimum likelihood) yöntemi ile bulunabilir. Bu durumda posterior olasılık fonksiyonlarını oluşturan parametreler öğrenme verisinin parametreler ışığında olabilirliğini (likelihood) en büyük değere ulaştıracak şekilde seçilir. Elimizdeki öğrenme verisinin $(y_i, x_i)$ şeklinde verildiğini düşünelim. Bu durumda likelihood fonksiyonu:

$$
L(\theta) = \prod_{i=1}^Np_{y_i}(x_i|\theta)
$$

şeklinde yazılabilir. Bu denklem belli bir $\theta$ değeri için öğrenme verisindeki her bir ölçümün sınıfı tahmin etme olasılıklarının çarpımı anlamına gelmektedir. Öğrenme verisinin bağımsiz ölçümlerden oluştuğunu düşünürsek bu olasılık değeri aslında parametreler ışığında öğrenme verisinin gözlenebilme (olabilirlik) olasılığını ifade edebilir. Yukarıdaki fonksiyonu maksimize etmek yerine genellikle bu fonksiyonun monoton transformasyonu (log likelihood) kullanılır. Bu durumda logistik regresyon parametreleri $\theta$:

$$
\hat{\theta} = \underset{\theta}{\mathrm{argmax}} \text{  } L(\theta) = \underset{\theta}{\mathrm{argmax}} \text{  } \text{log }(L(\theta)) = \sum_{i=1}^N \text{log } p_{y_i}(x_i|\theta)
$$
şeklinde bulunur.


Örneğin ikili sınıflama problemlerinde, daha önce belirttiğimiz gibi

$$
p_1(x|\theta) =  \frac{e^{\beta_{10} + \beta_1^Tx}}{1+e^{\beta_{10} + \beta_1^Tx}} = 
$$

ve

$$
p_0(x|\theta) =  \frac{1}{1+e^{\beta_{10} + \beta_1^Tx}} = 1-p_1(x|\theta)
$$

şeklinde ifade edilsin.


Bu durumda log olabilirlik (lok likelihood):

$$
\text{log }(L(\theta)) = \sum_{i=1}^N \left \{ y_i \text{log } p_1(x_i|\theta) + (1-y_i)\text{log }(1-p_1(x_i|\theta)) \right \}
$$


Yukaridak ifadede $y_i =1$ olduğunda $p_1(x)$, $y_i = 0$ olduğunda $p_0(x)$ olasılık fonksiyonlarının toplama dahil olduğunu görün. Bu ifade logaritma transformasyonu sonucunda aşağıdaki gibi yazılabilir:

$$
\text{log }(L(\theta)) = \sum_{i=1}^N \left \{ y_i\beta^Tx_i - \text{log }(1+e^{\beta^Tx_i} \right \}
$$

Burada $\beta = [\beta_{10}, \beta_1]$ şeklinde yazılır. Bu fonksiyonun maksimum değeri yukaridaki ifadenin türevinin sıfır yapan $\beta$ değerleri ile bulunabilir. Bu durumda türevi sıfıra eşitlersek:

$$
\nabla{\text{log } L(\beta)}=\frac{\partial{ \text{log }(L(\theta))}}{\partial{\beta}} = \sum_{i=1}^N x_i (y_i - p(x_i|\beta)) = 0
$$


Bu denklemin köklerini bulmak için Newton-Raphson iterasyonu kullanılabilir. Bu durumda log olabilirlik fonksiyonunun ikinci türevi (Hessian) ihtiyacımız vardır. Hessian matrisi:

$$
H(\beta) = \frac{\partial^2{ \text{log }(L(\theta))}}{\partial{\beta}\partial{\beta}^T} = -\sum_{i=1}^N x_i x_i^T p(x_i|\beta)(1-p(x_i|\beta))
$$

Daha önce kök bulmada kullandığımız gibi bir başlangıç tahmini $\beta_i, i=0$ ile Newton iterasyonu uygular isek $\beta_{i+1}$ şu şekilde yazılabilir.

$$
\beta_{i+1} = \beta_i - H(\beta_i)^{-1}\nabla(\text{log } L(\beta_i))
$$


Bu denklemleri matris notasyonu ile yazmak istersek, tüm öznitelikler ($x_i$) için $\beta_i$ kullanılarak elde edilen olasılık değerleri bir $p$ vektöründe toplanabilir.

$$
p = \begin{bmatrix}
p_1(x_1|\beta_i)\\
p_1(x_2|\beta_i)\\
\vdots\\
p_1(x_N|\beta_i)\\
\end{bmatrix}
$$

Öğrenme verisininde $x_i$ özniteliklerine karşılık gelen sınıf etiketi (0 veya 1) değerleri $y$ vektöründe toplansın.

$$
y = \begin{bmatrix}
y_1\\
y_2\\
\vdots\\
y_N\\
\end{bmatrix}
$$


öznitelikleri ($x_i$) ve $\beta_{10}$ parametresine denk gelen ve 1'lerden oluşan sütün vektörünü barındıran $X$ matrisi aşağıdaki gibi yazılsın. Bu durumda Hessian ve türev denklemi aşağıdaki gibi matrix ifadeleri olarak yazılabilir.

$$
\nabla log L(\beta_i) = X^T(y-p)
$$

ve

$$
H(\beta_i) = -X^TWX
$$

Burada $W$ diyagonal bir ağırlık matrisidir ve aşağıdaki gibi yazılabilir.

$$
W = diag(p(x_i|\beta_i)(1− p(x_i| \beta_i)) \text{,  } i \in 1,2,3,...,N)
$$


Bu durumda bir newton adımı aşağıdaki gibi yazılabilir

$$
\begin{array} {lcl} 
\beta_{i+1} & = & \beta_i + (X^TWX)^{-1}X^T(y-p) \\ 
&             = &  (X^TWX)^{-1}X^TW(X\beta_i + W^{-1}(y-p))\\
&             = & (X^TWX)^{-1}X^TWz
\end{array}
$$

şeklinde yazılabilir. Burada $z = X\beta_i + W^{-1}(y-p)$, düzeltilmiş çıktı (adjusted response olarak adlandırılabilir). İterasyona bakıldığında her adımda $W$, $p$ ve $z$ değerlerinin yeni parametrelere göre değiştiğini görebiliriz. Bu yöntem genel olarak Tekrarlı Ağırlıklandırmalı En Küçük Kareler (Iteratively Reweighted Least Squares - IRLS) olarak da bilinir. Birden fazla sınıf olduğunda benzeri şekilde IRLS kullanılabilir ancak yukarıdaki ifadeler uygun şekilde düzenlenmelidir. Genellikle $\beta_0 = 0$ uygun bir başlangıç değeri oluşturur.

```python
from matplotlib import animation, rc
import time
rc('animation', html='html5')

fig, ax = plt.subplots(figsize=(7,7))

ax.set_ylabel('Income')
ax.set_xlabel('Balance')


ax.scatter(df[df.default2==1].balance, df[df.default2==1].income, s=50, c='orange', marker='+',
            linewidths=1,alpha=0.5)
ax.scatter(df[df.default2==0].balance, df[df.default2==0].income, s=50, marker='o', linewidths='1',
            edgecolors='lightblue', facecolors='white', alpha=.5)

line, = ax.plot([], [], color='red', lw=2)

Y = df.default2.to_numpy()

X = np.vstack(
(
    np.ones(len(Y)),
    df.balance.to_numpy(),
    df.income.to_numpy()
)
).T

## beta 0 parametreleri 0'a eşitliyoruz
beta_0 = np.array([-1, -5.272e-04, 1.8e-06])


def prob(x, beta):
    ex = np.exp(np.dot(beta,x))
    return ex / (1+ex)

def pvec(X, beta):
    return np.array([prob(X[i,:],beta) for i in range(X.shape[0])])

#Calculate W and inverse of it
def calcW_and_inv_W(X, beta):
    px = pvec(X,beta)
    w = px*(1-px)
    W = np.diag(w)
    invW = np.diag(1/w)
    return W, invW

def log_regression(y, X, beta):
    nbeta = beta
    i=0
    while True:
        p = pvec(X,nbeta)
        W,invW = calcW_and_inv_W(X, nbeta)
        z = np.dot(X, nbeta) + np.dot(invW, y-p)
        XwX = np.dot(X.T, np.dot(W, X))
        wy = np.dot(X.T, np.dot(W, z))
        
        b = np.linalg.lstsq(XwX , wy,rcond=4)[0]
        incr = b - nbeta
        nbeta = b
        if (np.linalg.norm(incr) < 1e-8):
            nbeta = b
            break
        nbeta = b
        i=i+1
        yield i,b

        
    yield i,nbeta


param_iter = log_regression(y,X, beta_0)

def animate(i):
    b = beta_0
    k=0
    if (i!=0):
        k,b = next(param_iter)
    print ('Iteration ', k, ' Parameters ',b)
    x = np.arange(df.balance.min(), df.balance.max(), 100)
    #
    # Karar siniri beta^Tx = 0
    # Bu durumda Income = (-beta_0  - beta_1 * Balance)/beta_2
    # 
    y = -(b[0] + b[1]*x)/b[2]
    
    line.set_data(x, y)
    
    return (line,)    


clf = skl_lm.LogisticRegression(solver='newton-cg')
clf.fit(X,y)

print ('Parameter Scikit Learn ', clf.coef_)
anim = animation.FuncAnimation(fig, animate, init_func=None,
                               frames=10, interval=500, blit=True)


anim
```

## Lineer Diskriminant Analizi (LDA)

Sınıfların indikatör fonksiyonlarını posterior olasılıklarına bağlamanın lineer regresyona göre daha iyi sonuçlar verebildiğini bir önceki örneklerimizde görmüştük. Eğer girdi değerlerine göre sınıf posterior olasılığını Bayes teoremine göre yazarsak. Sınıfların koşullu olasılığını aşağıdaki şekilde ifade edebiliriz.

$$
d_k(x) = P(C = k| X = x) = \frac{P(X=x| C=k) P_k}{\sum_{l=1}^K P(X=x|C=l) P_l } = \frac{f_k(x) \pi_k}{\sum_{l=1}^K f_l(x) \pi_l }
$$


Bu ifadede $P_k$ k sınıfının ortaya çıkma olasılığını (class prior), $f_k(x) = P(X = x | C = k)$, k sınıfı olduğu bilindiğinde bu gözlemin özniteliğinin $x$ olma olasılığını (sınıf densite fonksiyonu), daha önce belirttiğimiz gibi $P(C = k| X = x)$ ise $x$ özniteliği gözlendiğinde bu gözlemin k sınıfında bulunma olasılığını ifade etmektedir. Literatürde sınıf ön bilgisi (prior) $P_k = \pi_k$ olarak ifade edilmektedir. Burada eğer $\pi_k$ ve densite fonksiyonu $f_k(x)$ kestirilebilirse yukarıdaki şekilde ifade edilen posterior sınıf olasılıklarını kullanarak bir Bayes sınıflandırıcısı elde edilebilir. Daha önce de belirttiğimiz gibi en büyük $d_k(x)$ değerine sınıflama yapan bir bayes sınıflandırıcısı tüm sınıflandırma yöntemlerine göre en düşük hata miktarına sahiptir. Bu durumda $f_k(x)$ densite fonksiyonu kestirilebilir ise yukarıdaki ifadeler kullanılarak yaklaşık bir Bayes sınıflandırıcısı oluşturulabilir. Lineer ve quadratik diskriminant analizleri $f_k(x)$ için Gauss Normal Dağılımını kullanmaktadır. Daha esnek yaklaşımlar Gauss karışımlarını (Mixture of Gaussian) kullanabilir. Parametrik olmayan yöntemler ile densite fonksiyonlarının kestirimi en geniş esnekliği sağlar.


Lineer diskriminant analizinde densite fonksiyonları çok değişenli Gauss dağılımı olarak ifade edilir.

$$
f_k(x) = \frac{1}{(2\pi)^{\frac{p}{2}}\sqrt{det(\Sigma)}}exp(-\frac{1}{2}(x-\mu_k)^T\Sigma^{-1}(x-\mu_k))
$$

Eğer tüm sınıfların aynı covaryans $\Sigma$ değerine sahip olduğu varsayarsak. Bu durumda $k$ ve $l$ sınıfları arasındaki Karar Sınırı monoton transform altında şu şekilde yazılabilir.

$$
log \frac{P(C=k|X=x)}{P(C=l|X=x)} = \text{log} \frac {f_k(x)}{f_l(x)} + \text{log} \frac {\pi_k}{\pi_l}
$$

yukaridaki multivariate Gauss dağılımını yerine yazarsak ve sadeleştirisek

$$
log \frac{P(C=k|X=x)}{P(C=l|X=x)} = \text{log} \frac {\pi_k}{\pi_l} -\frac{1}{2}(\mu_k+\mu_l)^T\Sigma^{-1}(\mu_k-\mu_l) + x^T\Sigma^{-1}(\mu_k-\mu_l)
$$

Yukaridaki ifadede $x$ öznitelik değerinin lineer fonksiyonu olduğunu görün. Bu durumda Lineer Diskriminant analizi ile doğrusal Karar Sınırları bulunacaktır. Yukarıdaki ifadede eğer sınıfların densite dağılım fonksiyonlarının ortalama değerleri $\mu_k$, covaryans matrisi $\Sigma$ ve prior olasılıkları $\pi_k$ kestirilebilir ise indikatör fonksiyonları aşağıdaki gibi yazılabilir.

$$
d_k(x) = x^T\Sigma^{-1}\mu_k  - \frac{1}{2}\mu_k\Sigma^{-1}\mu_k + \text{log} \pi_k
$$

Aşağıdaki şekilde üç sınıf için dağılım fonksiyonları ve %95 olasılık elipsleri ve bayes karar sınırları verilmiştir.Sağda ise her sınıftan 30 örnek alındığındaki realizasyon gösterilmiştir. Yukarıdaki ifadeyi logistik regresyon ile karşılaştırın ve logistik regresyon parametrelerini eşleştirmeye çalışın.

### Lineer Diskriminant Analizi Kestirimi 

LDA Gauss dağılımı parametrelerini oluşturmak için için öğrenme verisinden bu parametrelerin ($\hat{\pi_k}, \hat{\mu_k}, \hat{\Sigma}$) kestirilmesi gerekir. Bu kestirimler aşağıdaki gibi yapılabilir.

1. $\hat{\pi_k} = \frac{N_k}{N}$
2. $\hat{\mu_k} = \sum_{y_i = k} \frac{x_i}{N_k}$
3. $\hat{\Sigma} = \frac{1}{N-k} \sum_{k=1}^K\sum_{y_i = k} (x_i - \mu_k)(x_i - \mu_k)^T$

Burada $N_k$ k sınıfında bulunan gözlem sayısını, $N$ toplam gözlem sayısını vermektedir. Logistik regression için kullandığımız simülasyona benzer bir simülasyonu LDA ile kestirmeyi deneyelim.

```python
import scipy

def sample_class(mean, cov, size):

    class1 = scipy.stats.multivariate_normal(mean, cov)
    
    return class1.rvs(size=size)

cov  = [[100, 0], [0, 100]]

pos_class1 = sample_class([-8, -14], cov, 100)
pos_class2 = sample_class([12, 10], cov, 1000)


X1 = np.hstack([pos_class1[:,0],pos_class2[:,0]] )
X2 = np.hstack([pos_class1[:,1],pos_class2[:,1]] )

Y = np.hstack([np.ones(pos_class1.shape[0]),np.zeros(pos_class2.shape[0])])

X = np.vstack(
(
    X1,
    X2
)).T


def LDA(Y, X):
    pi_0 = len(Y[Y==0]) / len(Y)
    pi_1 = len(Y[Y==1])/ len(Y)
    
    mu_0 = np.mean(X[Y==0,:],axis=0)
    mu_1 = np.mean(X[Y==1,:],axis=0)
    
    sigma = (np.cov((X[Y==0,:]-mu_0).T) + np.cov((X[Y==1,:]-mu_1).T))

    return pi_0, pi_1, mu_0, mu_1, sigma
    
pi_0,pi_1,mu_0, mu_1,sigma = LDA(Y,X)

fig, ax = plt.subplots()

ax.set_ylabel('X2')
ax.set_xlabel('X1')


ax.scatter(X1[Y==1], X2[Y==1], s=50, c='orange', marker='+',
            linewidths=1,alpha=0.5)
ax.scatter(X1[Y==0], X2[Y==0], s=50, marker='o', linewidths='1',
            edgecolors='lightblue', facecolors='white', alpha=.5)



x = np.arange(X1.min(), X1.max(), (X1.max()- X1.min())/100)

#
# Karar siniri beta_0 + beta^Tx = 0
# LDA icin beta0 = log(pi1/pi0) - 1/2(m_k+m_l)^T inv (sigma) (m_k - m_l)
# beta = inv sigma (m_k - m_l)
#

## Note normally it is not convenient to work with matrix inverses, use linalg solve instead.
inv_sigma = np.linalg.inv(sigma)

b = np.dot(np.linalg.inv(sigma), mu_1-mu_0)
b0 = np.log(pi_1/pi_0) - 1/2.0 * np.dot(mu_1+mu_0, np.dot(inv_sigma, mu_1-mu_0))


if (b[1] != 0):
    y = -(b0 + b[0]*x)/b[1]
else:
    y = x
    x = np.ones_like(x)*(-b0/b[0])

line, = ax.plot(x, y, color='red', lw=2)


```

## Quadratik Discriminant Analizi QDA

LDA'den farkli olarak QDA her sınıfın kovaryans matrislerinin farklı olduğunu varsayar. Böylece kestirilecek parametre sayısı LDA'ye göre artmış olur, ancak karar sınırları daha esnek modellenebilir.  Bu durumda indikatör fonksiyonu

$$
d_k(x) =  -\frac{1}{2}\text{log det}(\Sigma_k)  - \frac{1}{2}(x-\mu_k)^T\Sigma_k^{-1}(x-\mu_k) + \text{log} \pi_k
$$

olarak yazilir. Burada ortadaki terimden dolayi diskriminant fonksiyonu girdi değerlerine quadratik olarak bağlıdır ($x^Tx = \sum x_i^2$).

```python
import scipy

np.random.seed(9999)
def sample_class(mean, cov, size):

    class1 = scipy.stats.multivariate_normal(mean, cov)
    
    return class1.rvs(size=size)

cov1  = [[100, 0], [0, 100]]

cov2  = [[60, 30], [30, 80]]


pos_class1 = sample_class([0, -10], cov1, 100)
pos_class2 = sample_class([12, 10], cov2, 1000)


X1 = np.hstack([pos_class1[:,0],pos_class2[:,0]] )
X2 = np.hstack([pos_class1[:,1],pos_class2[:,1]] )

Y = np.hstack([np.ones(pos_class1.shape[0]),np.zeros(pos_class2.shape[0])])

X = np.vstack(
(
    X1,
    X2
)).T


def QDA(Y, X):
    pi_0 = len(Y[Y==0]) / len(Y)
    pi_1 = len(Y[Y==1])/ len(Y)
    
    mu_0 = np.mean(X[Y==0,:],axis=0)
    mu_1 = np.mean(X[Y==1,:],axis=0)
    
    sigma_0 = np.cov((X[Y==0,:]-mu_0).T)
    sigma_1 = np.cov((X[Y==1,:]-mu_1).T)


    return pi_0, pi_1, mu_0, mu_1, sigma_0, sigma_1
               
               
def discr(x, pi, mu, sigma):
    det = np.linalg.det(sigma)
    xm = x-mu
    x1 = np.dot( np.linalg.inv(sigma), (x-mu).T).T
    x = np.array([np.dot(xm[i,:], x1[i,:]) for i in range(x1.shape[0])])
    
    return np.log(det) - 1/2.0*x + np.log(pi)


pi_0,pi_1,mu_0, mu_1,sigma_0,sigma_1 = QDA(Y,X)



Y_0_p = discr(X, pi_0, mu_0, sigma_0)
Y_1_p = discr(X, pi_1, mu_1, sigma_1)

Y_predicted = np.argmax(
np.vstack(
(Y_0_p,
 Y_1_p
)
),axis=0
)






print (pi_0,pi_1,mu_0, mu_1,sigma_0,sigma_1 )

fig, ax = plt.subplots()

ax.set_ylabel('X2')
ax.set_xlabel('X1')


ax.scatter(X1[Y==1], X2[Y==1], s=50, c='orange', marker='+',
            linewidths=1,alpha=0.5)
ax.scatter(X1[Y==0], X2[Y==0], s=50, marker='o', linewidths='1',
            edgecolors='lightblue', facecolors='white', alpha=.5)



nx, ny = 200, 100
x_min, x_max = plt.xlim()
y_min, y_max = plt.ylim()
xx, yy = np.meshgrid(np.linspace(x_min, x_max, nx),
                     np.linspace(y_min, y_max, ny))

X_mesh = np.vstack(
    ( xx.ravel(), yy.ravel())
).T
               
Z_1 = discr(X_mesh, pi_1, mu_1, sigma_1)
Z_0 = discr(X_mesh, pi_0, mu_0, sigma_0)





# En büyük diskriminant değerine sınıf ataması yap.
C = np.argmax(
np.vstack(
(Z_0,
 Z_1
)
),axis=0
)
plt.contour(xx, yy, C.reshape(ny,nx), linewidths=2., colors='red')



tp = np.sum((Y == 1) & (Y_predicted==1))
tn = np.sum((Y == 0) & (Y_predicted==0))
fp = np.sum((Y == 0) & (Y_predicted==1))
fn = np.sum((Y == 1) & (Y_predicted==0))



print ('Total :', len(Y), 'True positive:',tp, 'True Negative :',tn, 'False Positive :',fp, ' False Negative :',fn )
```
